```python
import tiktoken
enc = tiktoken.get_encoding("cl100k_base")
```

```python
# Count tokens for user column
data["user_token_count"] = data["user"].astype(str).apply(
    lambda x: len(enc.encode(x))
)
# Count tokens for chatbot column
data["chatbot_token_count"] = data["chatbot"].astype(str).apply(
    lambda x: len(enc.encode(x))
)
# Total tokens per row
data["total_tokens"] = data["user_token_count"] + data["chatbot_token_count"]
```

```python
data
```

```python
data['exceed_limit'] = data['total_tokens'] > 4096
```
